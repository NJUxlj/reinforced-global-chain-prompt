```python
import numpy as np  
import matplotlib.pyplot as plt  

#   
x = np.array([1e8, 1e9, 1e10])  
data1 = {  
    '1%': [47, 47, 82],  
    '5%': [51, 61, 89],  
    '10%': [57, 73, 88],  
    '20%': [59, 76, 90],  
    '50%': [58, 77, 87]  
}  

#   
plt.figure(figsize=(15, 10))  
# 2x2  
ax1 = plt.subplot(2, 2, 1)  
ax2 = plt.subplot(2, 2, 2)  
ax3 = plt.subplot(2, 2, 3)  
ax4 = plt.subplot(2, 2, 4) 

#   
markers = ['s', '+', 'o', 'x', 'd']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  
for (label, values), marker, color in zip(data1.items(), markers, colors):  
    ax1.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  

ax1.set_xscale('log')  
ax1.grid(True, alpha=0.3)  
ax1.set_xlabel('Model Parameters')  
ax1.set_ylabel('SuperGLUE Score')  
ax1.set_ylim(40, 100)  
ax1.legend()  
#  (a)   
# ax1.text(-0.1, 1.1, '(a) Prompt length', transform=ax1.transAxes)

#   
data2 = {  
    'TF-IQF-AS': [56, 67, 90],  
    'Clustering-based': [59, 74, 88],  
    'LDA': [60, 76, 90]  
}  

markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data2.items(), markers, colors):  
    ax2.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax2.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)  

ax2.set_xscale('log')  
ax2.grid(True, alpha=0.3)  
ax2.set_xlabel('Model Parameters')  
ax2.set_ylabel('Average MCQ Score')  
ax2.set_ylim(40, 100)  
ax2.legend()  




data3 = {  
    'MLP': [56, 67, 90],  
    'LSTM': [59, 74, 88],  
    'BPA-Encoder': [60, 76, 90]  
} 
markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data3.items(), markers, colors):  
    ax3.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax3.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)
ax3.set_xscale('log')  
ax3.grid(True, alpha=0.3)  
ax3.set_xlabel('Model Parameters')  
ax3.set_ylabel('Average MCQ Score')  
ax3.set_ylim(40, 100)  
ax3.legend()



data4 = {  
    'Input-Only': [56, 67, 90],  
    'Append-All-Layers': [59, 74, 88],  
    'Inject-All-Layers': [60, 76, 90]  
} 

markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data4.items(), markers, colors):  
    ax4.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax4.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)
ax4.set_xscale('log')  
ax4.grid(True, alpha=0.3)  
ax4.set_xlabel('Model Parameters')  
ax4.set_ylabel('Average MCQ Score')  
ax4.set_ylim(40, 100)  
ax4.legend()

ax4 = plt.subplot(2, 2, 4)

#   
plt.tight_layout()  
plt.show()
```
