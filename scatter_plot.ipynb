```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以保证结果可重复
np.random.seed(42)

# 创建两个子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# 左图数据 (a)
n_points = 50  # 每个簇的大概点数

# 生成四个簇的数据
cluster1 = np.random.normal(loc=(2, 1), scale=0.3, size=(n_points, 2))  # 蓝色簇
cluster2 = np.random.normal(loc=(1, 0), scale=0.2, size=(n_points, 2))  # 橙色簇
cluster3 = np.random.normal(loc=(3, 2), scale=0.3, size=(n_points, 2))  # 绿色簇
cluster4 = np.random.normal(loc=(4, 3), scale=0.3, size=(n_points, 2))  # 红色簇

# 绘制左图
ax1.scatter(cluster1[:, 0], cluster1[:, 1], c='blue', alpha=0.6)
ax1.scatter(cluster2[:, 0], cluster2[:, 1], c='orange', alpha=0.6)
ax1.scatter(cluster3[:, 0], cluster3[:, 1], c='green', alpha=0.6)
ax1.scatter(cluster4[:, 0], cluster4[:, 1], c='red', alpha=0.6)

# 右图数据 (b)
n_points_b = 30  # 每个簇的点数稍少

# 生成四个较小的簇
cluster1_b = np.random.normal(loc=(2, 3), scale=0.2, size=(n_points_b, 2))  # 蓝色簇
cluster2_b = np.random.normal(loc=(4, 2), scale=0.2, size=(n_points_b, 2))  # 橙色簇
cluster3_b = np.random.normal(loc=(2, 2), scale=0.2, size=(n_points_b, 2))  # 绿色簇
cluster4_b = np.random.normal(loc=(4, 3), scale=0.2, size=(n_points_b, 2))  # 红色簇

# 绘制右图
ax2.scatter(cluster1_b[:, 0], cluster1_b[:, 1], c='blue', alpha=0.6)
ax2.scatter(cluster2_b[:, 0], cluster2_b[:, 1], c='orange', alpha=0.6)
ax2.scatter(cluster3_b[:, 0], cluster3_b[:, 1], c='green', alpha=0.6)
ax2.scatter(cluster4_b[:, 0], cluster4_b[:, 1], c='red', alpha=0.6)

# 设置图表属性
ax1.set_title('(a)')
ax2.set_title('(b)')

# 设置坐标轴范围和网格
for ax in [ax1, ax2]:
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_aspect('equal')

# 调整子图之间的间距
plt.tight_layout()

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

# 设置seaborn样式
sns.set_theme(style="whitegrid")
sns.set_context("paper")

# 设置全局字体和线条样式
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 10
rcParams['axes.linewidth'] = 0.8
rcParams['xtick.major.width'] = 0.8
rcParams['ytick.major.width'] = 0.8
rcParams['xtick.direction'] = 'in'
rcParams['ytick.direction'] = 'in'

# 使用学术论文友好的配色方案
colors = sns.color_palette('Set1', n_colors=4)

# 设置随机种子
np.random.seed(42)

# 创建图表，使用黄金比例
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), dpi=300)

# 左图数据 (a)
n_points = 50
clusters_a = [
    np.random.normal(loc=(2, 1), scale=0.3, size=(n_points, 2)),
    np.random.normal(loc=(1, 0), scale=0.2, size=(n_points, 2)),
    np.random.normal(loc=(3, 2), scale=0.3, size=(n_points, 2)),
    np.random.normal(loc=(4, 3), scale=0.3, size=(n_points, 2))
]

# 右图数据 (b)
n_points_b = 30
clusters_b = [
    np.random.normal(loc=(2, 3), scale=0.2, size=(n_points_b, 2)),
    np.random.normal(loc=(4, 2), scale=0.2, size=(n_points_b, 2)),
    np.random.normal(loc=(2, 2), scale=0.2, size=(n_points_b, 2)),
    np.random.normal(loc=(4, 3), scale=0.2, size=(n_points_b, 2))
]

# 绘制函数
def plot_clusters(ax, clusters, labels=None):
    for i, cluster in enumerate(clusters):
        ax.scatter(cluster[:, 0], cluster[:, 1], 
                  c=[colors[i]], 
                  s=30,  # 点的大小
                  alpha=0.7,  # 透明度
                  label=f'Cluster {i+1}' if labels else None,
                  edgecolors='none')  # 移除边框
    
    # 设置坐标轴
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(True, linestyle='--', alpha=0.3, color='gray')
    ax.set_aspect('equal')
    
    # 设置刻度
    ax.tick_params(axis='both', which='major', labelsize=9)
    
    # 添加轻微的内边距
    ax.margins(0.1)

# 绘制两个子图
plot_clusters(ax1, clusters_a, labels=True)
plot_clusters(ax2, clusters_b)

# 设置标题
ax1.set_title('(a)', pad=10, fontsize=11, fontweight='bold')
ax2.set_title('(b)', pad=10, fontsize=11, fontweight='bold')

# 只在左图显示图例
legend = ax1.legend(frameon=True, 
                   fancybox=False, 
                   edgecolor='black',
                   fontsize=8,
                   bbox_to_anchor=(1.05, 1),
                   loc='upper left')
legend.get_frame().set_linewidth(0.5)

# 调整布局
plt.tight_layout()

# 保存图片，使用高DPI确保清晰度
try:
    plt.savefig('clustering_plot.pdf', 
                dpi=300, 
                bbox_inches='tight',
                pad_inches=0.1)
    plt.savefig('clustering_plot.png', 
                dpi=300, 
                bbox_inches='tight',
                pad_inches=0.1)
except Exception as e:
    print(f"保存文件时出错: {e}")
    print("继续显示图表...")

plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  

# 设置现代化的图表风格  
sns.set_theme(style="white", context="paper")  # 使用纯白背景  
plt.rcParams.update({  
    'font.family': 'Arial',  
    'font.size': 12,  
    'axes.linewidth': 1.0,  
    'axes.labelsize': 14,  
    'axes.spines.top': True,  
    'axes.spines.right': True,  
    'axes.spines.left': True,  
    'axes.spines.bottom': True,  
    'figure.dpi': 300,  
    'savefig.dpi': 300,  
})  

# 设置随机种子以确保可重复性  
np.random.seed(42)  

# 生成两组数据点（减小标准差使点更密集）  
# 第一组（蓝色）  
x1 = np.concatenate([  
    np.random.normal(2, 0.3, 50),  # 减小标准差  
    np.random.normal(4, 0.4, 50),  
    np.random.normal(6, 0.3, 50)  
])  
y1 = np.concatenate([  
    np.random.normal(2, 0.3, 50),  
    np.random.normal(4, 0.4, 50),  
    np.random.normal(6, 0.3, 50)  
])  

# 第二组（红色）  
x2 = np.concatenate([  
    np.random.normal(3, 0.3, 40),  
    np.random.normal(5, 0.4, 40),  
    np.random.normal(8, 0.3, 40)  
])  
y2 = np.concatenate([  
    np.random.normal(3, 0.3, 40),  
    np.random.normal(5, 0.4, 40),  
    np.random.normal(3, 0.3, 40)  
])  

# 创建图形  
fig, ax = plt.subplots(figsize=(8, 6))  

# 绘制散点图  
scatter1 = ax.scatter(x1, y1, c='blue', s=50, alpha=0.6, label='Group A',   
                     edgecolor='none')  
scatter2 = ax.scatter(x2, y2, c='red', s=50, alpha=0.6, label='Group B',  
                     edgecolor='none')  

# 移除刻度  
ax.set_xticks([])  
ax.set_yticks([])  

# 设置所有边框可见且为黑色  
for spine in ax.spines.values():  
    spine.set_visible(True)  
    spine.set_color('black')  
    spine.set_linewidth(1.0)  

# 调整布局  
plt.tight_layout()  

# 保存图形  
plt.savefig('scatter_plot.pdf', bbox_inches='tight', facecolor='white')  
plt.savefig('scatter_plot.png', bbox_inches='tight', facecolor='white')  

# 显示图形  
plt.show()
```
