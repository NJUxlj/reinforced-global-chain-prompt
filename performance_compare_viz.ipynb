```python
import matplotlib.pyplot as plt  
import numpy as np  

# 数据准备  
labels = ['BaasPrompt\n(base-scale ~110M)', 'P-tuning\n(base-scale ~110M)',   
          'Baas-Prompt\n(large-scale ~340M)', 'P-tuning\n(large-scale ~340M)']  
bert_scores = [66, 68, 72.5, 74.5]  
gpt_scores = [63, 70, 63, 77.6]  

# 设置图形大小和样式  
plt.figure(figsize=(12, 6))  
plt.style.use('seaborn-v0_8-whitegrid')  

# 设置柱状图的位置  
x = np.arange(len(labels))  
width = 0.35  

# 创建柱状图  
rects1 = plt.bar(x - width/2, bert_scores, width, label='BERT', color='#4285f4')  
rects2 = plt.bar(x + width/2, gpt_scores, width, label='Roberta', color='#ffa726')  

# 设置标签和标题  
plt.ylabel('MCQA average')  
plt.xticks(x, labels)  
plt.legend()  

# 设置y轴范围和网格样式  
plt.ylim(60, 80)  
plt.grid(axis='y', linestyle='--', alpha=0.3)  

# 添加水平虚线连接和差值标注  
# 第一组差值  
plt.hlines(y=66, xmin=0-width/2, xmax=1-width/2, colors='gray', linestyles='--', alpha=0.5)  
plt.annotate('+4.2', xy=(1.1, 70.5), xytext=(1.1, 70.5), fontsize=10)  

# 第二组差值  
plt.hlines(y=74.5, xmin=2-width/2, xmax=3-width/2, colors='gray', linestyles='--', alpha=0.5)  
plt.annotate('+3.1', xy=(3.1, 75), xytext=(3.1, 75), fontsize=10)  

# 移除顶部和右侧边框  
plt.gca().spines['top'].set_visible(False)  
plt.gca().spines['right'].set_visible(False)  

# 调整布局  
plt.tight_layout()  

# 显示图形  
plt.show()
```
