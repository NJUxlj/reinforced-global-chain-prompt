```python
import numpy as np  
import matplotlib.pyplot as plt  

#   
# x = np.array([0.5e9, 1.5e9, 3e9, 7e9]) 
x = np.array([0, 1, 2, 3])
data1 = {  
    '1%': [47, 47, 82, 90],  
    '5%': [51, 61, 89, 92],  
    '10%': [57, 73, 88, 90],  
    '20%': [59, 76, 90, 93],  
    '50%': [58, 77, 87, 91]  
}  

#   
plt.figure(figsize=(15, 10))  
# 2x2  
ax1 = plt.subplot(2, 2, 1)  
ax2 = plt.subplot(2, 2, 2)  
ax3 = plt.subplot(2, 2, 3)  
ax4 = plt.subplot(2, 2, 4) 

# x  
def set_x_format(ax):  
    # ax.set_xscale('log')  
    ax.set_xticks(x)  #   
    ax.set_xticklabels(['124M', '335M', '774M', '1.5B'])  #   
    ax.grid(True, alpha=0.3)  
    ax.set_xlabel('Model Parameters')  
    ax.set_ylabel('Average MCQ F1')  
    ax.set_ylim(40, 100) 

#   
markers = ['s', '+', 'o', 'x', 'd']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']  
for (label, values), marker, color in zip(data1.items(), markers, colors):  
    ax1.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  

# ax1.set_xscale('log')  
ax1.grid(True, alpha=0.3)  
ax1.set_xlabel('Model Parameters')  
ax1.set_ylabel('Average MCQ F1')  
ax1.set_ylim(40, 100)  
ax1.legend()  
set_x_format(ax1)
#  (a)   
# ax1.text(-0.1, 1.1, '(a) Prompt length', transform=ax1.transAxes)

#   
data2 = {  
    'TF-IQF-AS': [56, 67, 90, 92],  
    'Clustering-based': [59, 74, 88, 90],  
    'LDA': [60, 76, 90, 93]  
}  

markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data2.items(), markers, colors):  
    ax2.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax2.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)  

# ax2.set_xscale('log')  
ax2.grid(True, alpha=0.3)  
ax2.set_xlabel('Model Parameters')  
ax2.set_ylabel('Average MCQ F1')  
ax2.set_ylim(40, 100)  
ax2.legend()  
set_x_format(ax2)



data3 = {  
    'MLP': [56, 67, 90, 91],  
    'LSTM': [59, 74, 88, 92],  
    'BPA-Encoder': [60, 76, 90, 93]  
} 
markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data3.items(), markers, colors):  
    ax3.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax3.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)
# ax3.set_xscale('log')  
ax3.grid(True, alpha=0.3)  
ax3.set_xlabel('Model Parameters')  
ax3.set_ylabel('Average MCQ F1')  
ax3.set_ylim(40, 100)  
ax3.legend()
set_x_format(ax3)


data4 = {  
    'Input-Only': [56, 67, 90, 91],  
    'Append-All-Layers': [59, 74, 88, 92],  
    'Inject-All-Layers': [60, 76, 90, 93]  
} 

markers = ['s', 'o', 'x']  
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  

for (label, values), marker, color in zip(data4.items(), markers, colors):  
    ax4.plot(x, values, label=label, marker=marker, linestyle='-', color=color)  
    #   
    ax4.fill_between(x, np.array(values)-2, np.array(values)+2, alpha=0.2, color=color)
# ax4.set_xscale('log')  
ax4.grid(True, alpha=0.3)  
ax4.set_xlabel('Model Parameters')  
ax4.set_ylabel('Average MCQ F1')  
ax4.set_ylim(40, 100)  
ax4.legend()
set_x_format(ax4)

ax4 = plt.subplot(2, 2, 4)

#   
plt.tight_layout()  
plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

#   
# (a) NLU: RACE  
# x1 = np.linspace(0, 80, 10)  
x1 = np.array([1, 5, 10, 20, 50])
embedding1 = np.array([87, 85, 86.2, 88.3, 87])  
mlp1 = np.array([88, 89, 87.5, 86, 87.3])  
bap1 = np.array([88, 89, 87.5, 86, 87.3]) 

# (b) NLU: SciQ  
x2 = np.array([1, 5, 10, 20, 50])
embedding2 = np.array([85.5, 86.5, 87.5, 87, 87])  
mlp2 = np.array([87, 89, 88, 88.7, 87.2])  
bap2 = np.array([88, 89, 87.5, 86, 87.3]) 

# (c) NLU: DREAM  
x3 = np.array([1, 5, 10, 20, 50]) 
embedding3 = np.array([83.7, 83.8, 83.5, 84.5, 83.6])  
mlp3 = np.array([83.5, 84.5, 83.7, 83.5, 84.3])  
bap3 = np.array([88, 89, 87.5, 86, 87.3]) 

# (d) NLU: CommonsenseQA  
x4 = np.array([1, 5, 10, 20, 50])
embedding4 = np.array([0.82, 0.83, 0.832, 0.835,  0.822])  
mlp4 = np.array([0.808, 0.81, 0.82, 0.822, 0.823]) 
bap4 = np.array([88, 89, 87.5, 86, 87.3]) 

#   
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))  

#   
def plot_subplot(ax, x, emb, mlp, bap, title, ylabel="Dev/f1"):  
    ax.plot(x, emb, '-o', label='embedding', color='#1f77b4')  
    ax.plot(x, mlp, '-o', label='MLP', color='#ff7f0e')  
    ax.plot(x, bap, '-*', label='BAP-Encoder', color='#008000') 
    ax.set_xlabel('')  
    ax.set_ylabel(ylabel)  
    ax.legend()  
    ax.grid(True, linestyle='--', alpha=0.7)  
    ax.set_title(title) 

    # x  
    xticks = ax.get_xticks()  
    #   
    ax.set_xticklabels([f'{int(x)}%' for x in xticks])  

plot_subplot(ax1, x1, embedding1, mlp1, bap1, "(a) RACE")  
plot_subplot(ax2, x2, embedding2, mlp2, bap2, "(b) SciQ")  
plot_subplot(ax3, x3, embedding3, mlp3, bap3, "(c) DREAM")  
plot_subplot(ax4, x4, embedding4, mlp4, bap4, "(d) CommonsenseQA")  

#   
plt.tight_layout()  
plt.show()
```
