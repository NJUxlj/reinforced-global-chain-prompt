```python
import numpy as np  
import matplotlib.pyplot as plt  

# GPT2-medium

#   
# (a) NLU: RACE  
# x1 = np.linspace(0, 80, 10)  
x1 = np.array([1, 5, 10, 20, 50])
embedding1 = np.array([87, 85, 86.2, 88.3, 87])  
mlp1 = np.array([88, 89, 87.5, 86, 87.3])  
bap1 = np.array([88, 89, 87.5, 86, 87.3]) 

# (b) NLU: SciQ  
x2 = np.array([1, 5, 10, 20, 50])
embedding2 = np.array([85.5, 86.5, 87.5, 87, 87])  
mlp2 = np.array([87, 89, 88, 88.7, 87.2])  
bap2 = np.array([88, 89, 87.5, 86, 87.3]) 

# (c) NLU: DREAM  
x3 = np.array([1, 5, 10, 20, 50]) 
embedding3 = np.array([83.7, 83.8, 83.5, 84.5, 83.6])  
mlp3 = np.array([83.5, 84.5, 83.7, 83.5, 84.3])  
bap3 = np.array([88, 89, 87.5, 86, 87.3]) 

# (d) NLU: CommonsenseQA  
x4 = np.array([1, 5, 10, 20, 50])
embedding4 = np.array([0.82, 0.83, 0.832, 0.835,  0.822])  
mlp4 = np.array([0.808, 0.81, 0.82, 0.822, 0.823]) 
bap4 = np.array([88, 89, 87.5, 86, 87.3]) 

#   
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))  

#   
def plot_subplot(ax, x, emb, mlp, bap, title, ylabel="Dev/f1"):  
    ax.plot(x, emb, '-o', label='embedding', color='#1f77b4')  
    ax.plot(x, mlp, '-o', label='MLP', color='#ff7f0e')  
    ax.plot(x, bap, '-*', label='BAP-Encoder', color='#008000') 
    ax.set_xlabel('')  
    ax.set_ylabel(ylabel)  
    ax.legend()  
    ax.grid(True, linestyle='--', alpha=0.7)  
    ax.set_title(title) 

    # x  
    xticks = ax.get_xticks()  
    #   
    ax.set_xticklabels([f'{int(x)}%' for x in xticks])  

plot_subplot(ax1, x1, embedding1, mlp1, bap1, "(a) RACE")  
plot_subplot(ax2, x2, embedding2, mlp2, bap2, "(b) SciQ")  
plot_subplot(ax3, x3, embedding3, mlp3, bap3, "(c) DREAM")  
plot_subplot(ax4, x4, embedding4, mlp4, bap4, "(d) CommonsenseQA")  

#   
plt.tight_layout()  
plt.show()
```
